```python
#get crypto data for last 5 years
def get_crypto_data(ticker):
    start = datetime(2017, 1, 1)
    end = datetime.now().date()
    df = web.DataReader(ticker, 'yahoo', start, end)
    return df
get_crypto_data('BTC-USD')
```

```python


def get_data(ticker):
    '''get last 5 years data from yahoo finance'''
    now = datetime.now()
    start = datetime(now.year - 5, now.month, now.day)
    return web.DataReader(ticker, 'yahoo', start=start, end=now)

get_data('AAPL')
```
